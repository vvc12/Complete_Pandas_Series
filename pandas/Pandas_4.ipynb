***this is another method for applying ffill() function either u can do upper one or below one***

```python
last_valid={
    'temperature':None,
    'windspeed':None,
    'event':None
}
def carry_forward_cell(cell,column_name):
    global last_valid
    if pd.isna(cell):
        return last_valid[column_name]
    else:
        last_valid[column_name]=cell
        return cell
for column in  ['temperature', 'windspeed', 'event']:
    df5[column]=df5[column].apply(lambda cell:carry_forward_cell(cell,column))
df5
```
