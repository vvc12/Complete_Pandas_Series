```python
import pandas as pd


data = {
    'day': ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'],
    'chicago': [32, 30, 28, 22, 30, 20, 25],
    'chennai': [75, 77, 75, 82, 83, 81, 77],
    'berlin': [41, 43, 45, 38, 30, 45, 47]
}
df = pd.DataFrame(data)
df
```

```python
df1=pd.melt(df,id_vars=["day"])#here id_vars is what must not be chnaged during data manipulaton i.e reshaping dataframe here melt is also an mehtod for reshaping like pivot table
df1
```

```python
#we can even filter the data like
df1[df1["variable"]=="chicago"]
```

```python
#but here column names are bad so we modify them using var_names
```

```python
df2=pd.melt(df,id_vars=["day"],var_name="city",value_name="temperature")
df2
```

```python
df2[df2["city"]=="chicago"]
```
