```python
import pandas as pd

# Sample DataFrame
data = {
    'temperature': ['23C', '25F', '30C', '21F','25F'],
    'windpseed': ['10mph', '12kmh', '15 mph', '5kmh', '50 mph'],
    'humidity': ['60%', '55%', '70%', '65%', '80%']
}

df6 = pd.DataFrame(data)

print("Original DataFrame:")
print(df6)

# Replace alphabetical characters (optionally preceded by spaces) in specified columns with an empty string
df7 = df6.replace({
    'temperature': '\s*[A-Za-z]',
    'windpseed': '\s*[A-Za-z]'
}, '', regex=True)

print("\nUpdated DataFrame:")
print(df7)
```
