```python
import pandas as pd
import mercury as mr
data = [
    ['Kathy', 'USA', 'Female', 23, 'Right'],
    ['Linda', 'USA', 'Female', 18, 'Right'],
    ['Peter', 'USA', 'Male', 19, 'Right'],
    ['John', 'USA', 'Male', 22, 'Left'],
    ['Fatima', 'Bangadesh', 'Female', 31, 'Left'],
    ['Kadir', 'Bangadesh', 'Male', 25, 'Left'],
    ['Dhaval', 'India', 'Male', 35, 'Left'],
    ['Sudhir', 'India', 'Male', 31, 'Left'],
    ['Parvir', 'India', 'Male', 37, 'Right'],
    ['Yan', 'China', 'Female', 52, 'Right'],
    ['Juan', 'China', 'Female', 58, 'Left'],
    ['Liang', 'China', 'Male', 43, 'Left']
]

df = pd.DataFrame(data, columns=['Name', 'Nationality', 'Sex', 'Age', 'Handedness'])
df
```

```python
df_styled = df.style.set_properties(**{'border': '1px solid white'}).set_table_styles(
    [
        {'selector': 'thead th', 'props': [('border', '1px solid white')]},
        {'selector': 'tbody th', 'props': [('border', '1px solid white')]}
    ]
)
df_styled
```

```python
pd.crosstab(df.Nationality,df.Handedness)
```

```python
#it shows the frequency
```

```python
pd.crosstab(df.Sex,df.Handedness)
```

```python
df1=pd.crosstab(df.Sex,df.Handedness,margins=True)
df1
```

```python
df2=pd.crosstab(df.Sex,[df.Nationality,df.Handedness],margins=True)
df2
```

```python
df3=pd.crosstab(df.Sex,df.Handedness,normalize="index")
df3
```

```python
import numpy as np
df4=pd.crosstab(df.Sex,df.Handedness,values=df.Age, aggfunc=np.average)
df4
```
