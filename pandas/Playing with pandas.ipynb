```python
%time
import random
import string 
import datetime
import pandas as pd
import numpy as np
# Generate 100,000 random integers
random_ints = [random.randint(1, 100) for _ in range(100_000)]

# Generate 100,000 random floating-point numbers
random_floats = [random.uniform(0.0, 10.0) for _ in range(100_000)]

# Generate 100,000 random boolean values
random_bools = [random.choice([True, False]) for _ in range(100_000)]

# Generate 100,000 random strings
random_strings = [''.join(random.choices(string.ascii_letters + string.digits, k=10)) for _ in range(100_000)]

# Generate 100,000 random dates
random_dates = [datetime.date(random.randint(2000, 2023), random.randint(1, 12), random.randint(1, 28)) for _ in range(100_000)]
df = pd.DataFrame({
    'integer': random_ints,
    'float': random_floats,
    'boolean': random_bools,
    'string': random_strings,
    'date': random_dates
})
```
